```python
import csv
```

```python
dataset=[]
with open('Order2.csv') as file:
    reader=csv.reader(file,delimiter=',')
    for row in reader:
        dataset+=[row]
```

```python
dataset[0:10]
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
#convert data to transaction encoder
te=TransactionEncoder()
x=te.fit_transform(dataset)
x
```

```python
#bulid model
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
dataset[0:10]
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(x,columns=te.columns_)
```

```python
df
```

```python
freq_itemset=apriori(df,min_support=0.01,use_colnames=True)
```

```python
freq_itemset
```

```python
rules=association_rules(freq_itemset,metrix='confidance',min_threshold=0.25)
```

```python
rules=association_rules(freq_itemset,metric='confidance',min_threshold=0.25)
```

```python
rules=association_rules(freq_itemset,metric='confidence',min_threshold=0.25)
```

```python
rules
```

```python
rules=rules[['antecedents','consequents','support','confidence']]
```
