```python
import csv
```

```python
dataset=[]
with open('Order1.csv') as file:
    reader=csv.reader(file,delimiter=',')
    for row in reader:
        dataset+=[row]
```

```python
dataset[0:10]
```

```python
#true false TransactionEncoder
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
te=TransactionEncoder()
x=te.fit_transform(dataset)
x
```

```python
#heading
import pandas as pd
```

```python
df=pd.DataFrame(x,columns=te.columns_)
```

```python
df.head()
```

```python
#find freq item
freq_item=apriori(df,min_support=0.01,use_colnames=True)
freq_item
```

```python
#find rules
rules=association_rules(freq_item,metric='confidence',min_threshold=0.05)
```
